The z-score approach (sometimes termed normalization) creates a distribution with a mean of 0 and a standard deviation of 1. Each standardized value is calculated by subtracting the feature mean and dividing by the quality deviation.

```python
# this data set is include just numeric features with multi labels 
#created for feature selection
# creating a dataframe with only numeric attributes  
numeric_col = X.select_dtypes(include='number').columns
numeric_data =X[numeric_col]
```

```python
# copy the data
df_z_scaled = numeric_data.copy()

# apply normalization techniques
for column in df_z_scaled.columns:
	df_z_scaled[column] = (df_z_scaled[column] -
						df_z_scaled[column].mean()) / df_z_scaled[column].std()	

# view normalized data
display(df_z_scaled)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from pyswarm import pso
```

```python

# Separate features and target
X1 = numeric_data

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.25, random_state=42)
```

```python
# Calculate Z-scores for all features in the training and testing sets
X_train_zscored = (X_train - X_train.mean(axis=0)) / X_train.std(axis=0)
X_test_zscored = (X_test - X_train.mean(axis=0)) / X_train.std(axis=0)
```

```python

# Define the classification model (you can replace this with any model of your choice)
clf = RandomForestClassifier(random_state=42)

# Define the fitness function using Z-scored features
def fitness_function(features):
    selected_features = np.where(features)[0]
    if len(selected_features) == 0:
        return 0.0  # Return a low fitness for no features selected
    
    # Train the classification model using Z-scored features
    clf.fit(X_train_zscored.iloc[:, selected_features], y_train)
    
    # Apply the same transformation to the test set
    X_test_selected = X_test_zscored.iloc[:, selected_features]
    
    # Make predictions on the test set
    y_pred = clf.predict(X_test_selected)
    
    # Calculate accuracy score (you can use other metrics as well)
    accuracy = accuracy_score(y_test, y_pred)
    
    return accuracy

# Define the bounds for each feature's selection status (0 for not selected, 1 for selected)
feature_count = X_train.shape[1]
lb = [0] * feature_count
ub = [1] * feature_count

# Run PSO to find the optimal feature subset
best_features, _ = pso(fitness_function, lb, ub, swarmsize=10, maxiter=50)

best_selected_indices = np.where(best_features)[0]

# Train and evaluate the final model using Z-scored features
X_test_selected = X_test_zscored.iloc[:, best_selected_indices]
y_pred = clf.predict(X_test_selected)
final_accuracy = accuracy_score(y_test, y_pred)

print(f"Selected features: {best_selected_indices}")
print(f"Final accuracy with selected features: {final_accuracy}")
```

# Building of the Reinforcement Model¶

```python
import torch
import torch.nn as nn
import torch.optim as optim
import random

# Define the DQN network
class DQN(nn.Module):
    def __init__(self, input_size, output_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(input_size, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, output_size)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)


X = selected_feature_Xtrain
y = y_train


# Convert data to PyTorch tensors
X_tensor = torch.FloatTensor(X.values)
y_tensor = torch.LongTensor(y.values)

# Hyperparameters
batch_size = 32
learning_rate = 0.001
gamma = 0.99
exploration_prob = 0.1
target_update_frequency = 10
num_episodes = 1000

# Initialize DQN and target network
input_size = X_tensor.shape[1]
output_size = 2  # Two actions: predict non-defect or defect
policy_net = DQN(input_size, output_size)
target_net = DQN(input_size, output_size)
target_net.load_state_dict(policy_net.state_dict())
target_net.eval()

# Define the loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(policy_net.parameters(), lr=learning_rate)

# Training loop
for episode in range(num_episodes):
    state = 0  # Start with the first data point
    total_reward = 0
    while state < len(X_tensor):
        # Epsilon-greedy exploration
        if np.random.uniform(0, 1) < exploration_prob:
            action = torch.tensor([random.randrange(output_size)], dtype=torch.long)
        else:
            with torch.no_grad():
                action = policy_net(X_tensor[state]).argmax(dim=0)

        # Simulate taking the action and receive reward
        reward = -1 if action != y_tensor[state] else 1
        total_reward += reward

        # Update Q-values
        next_state = state + 1
        if next_state >= len(X_tensor):
            break

        target = reward + gamma * target_net(X_tensor[next_state]).max().item()
        loss = criterion(policy_net(X_tensor[state])[action], torch.tensor([target]))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        state = next_state

    # Update target network every few episodes
    if episode % target_update_frequency == 0:
        target_net.load_state_dict(policy_net.state_dict())

    print(f"Episode {episode+1}, Total Reward: {total_reward}")

print("Training finished.")

# Testing
correct_predictions = 0
for i in range(len(X_tensor)):
    with torch.no_grad():
        action = policy_net(X_tensor[i]).argmax(dim=0)
    if action == y_tensor[i]:
        correct_predictions += 1

accuracy = correct_predictions / len(X_tensor)
print(f"Accuracy on test set: {accuracy:.2f}")
```
